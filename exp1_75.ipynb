```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")
```

```python
num = 7  # You can change this number

print(f"Multiplication Table for {num}:")
for i in range(1, 11):
   print(f"{num} x {i} = {num * i}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 12321
print(is_palindrome(num2))

num3 = 12345
print(is_palindrome(num3))

num4 = 0
print(is_palindrome(num4))
```

### Simple Expert System for Activity Recommendations

This expert system recommends activities based on current weather conditions (temperature, precipitation, wind, and cloud cover). It uses a set of predefined rules to make a decision.

```python
def recommend_activity(temperature, precipitation, wind, cloud_cover):
    """
    Recommends an activity based on weather conditions.

    Args:
        temperature (int): Temperature in Celsius.
        precipitation (str): Type of precipitation (e.g., 'None', 'Rain', 'Snow').
        wind (str): Wind strength (e.g., 'Calm', 'Breezy', 'Windy').
        cloud_cover (str): Cloud cover (e.g., 'Clear', 'Partly Cloudy', 'Overcast').

    Returns:
        str: Recommended activity.
    """

    if temperature >= 25 and precipitation == 'None' and wind == 'Calm' and cloud_cover == 'Clear':
        return "Go for a swim or sunbathe at the beach!"
    elif temperature >= 20 and precipitation == 'None' and (wind == 'Calm' or wind == 'Breezy') and (cloud_cover == 'Clear' or cloud_cover == 'Partly Cloudy'):
        return "Enjoy a hike or a picnic outdoors."
    elif temperature >= 15 and precipitation == 'None' and wind == 'Calm':
        return "Ideal weather for a leisurely walk in the park."
    elif temperature >= 10 and precipitation == 'None' and wind == 'Calm':
        return "Good for light outdoor activities, maybe a coffee on a patio."
    elif precipitation == 'Rain' or precipitation == 'Snow' or wind == 'Windy':
        if temperature < 5 and precipitation == 'Snow':
            return "Build a snowman or enjoy a warm drink indoors."
        else:
            return "Stay indoors: read a book, watch a movie, or play board games."
    elif cloud_cover == 'Overcast' and temperature < 15:
        return "Visit a museum or an indoor art gallery."
    else:
        return "Check local events; conditions are moderate for various options."
```

```python
# Example Usage:
print("--- Example 1 ---")
recommendation1 = recommend_activity(28, 'None', 'Calm', 'Clear')
print(f"Recommendation: {recommendation1}")

print("\n--- Example 2 ---")
recommendation2 = recommend_activity(10, 'Rain', 'Breezy', 'Overcast')
print(f"Recommendation: {recommendation2}")

print("\n--- Example 3 ---")
recommendation3 = recommend_activity(22, 'None', 'Breezy', 'Partly Cloudy')
print(f"Recommendation: {recommendation3}")

print("\n--- Example 4 ---")
recommendation4 = recommend_activity(-2, 'Snow', 'Calm', 'Overcast')
print(f"Recommendation: {recommendation4}")
```
